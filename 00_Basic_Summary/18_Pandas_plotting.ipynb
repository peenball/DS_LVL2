```python
import pandas as pd
```

```python
df = pd.DataFrame({"지점": ["A", "A", "B", "B"],
                   "상품": ["P1", "P2", "P1", "P2"],
                   "Q1": [112, 134, 156, 178],
                   "Q2": [212, 234, 256, 278],
                   "Q3": [312, 334, 356, 378],
                   "Q4": [412, 434, 456, 478]})
df
```

Pandas 객체는 .plot 접근자 또는 .plot() 메서드가 matplotlib 라이브러리로 구현할 수 있는 그래프를 손쉽게 그릴 수 있도록 지원함.

```python
df["Q1"].plot()
```

Series 객체에서 .plot() 메서드를 사용하는 경우 선그래프를 그려주며 x축은 인덱스, y축은 값(values)으로 자동 지정된다.

.plot() 메서드에서 그래프 종류를 변경하고자 할 때는 "kind" 인자에 별도의 값을 할당하여 원하는 그래프를 그릴 수 있다.

```python
df["Q1"].plot(kind = "bar")
```

```python
# 각 인덱스와 값이 서로 매칭되고 값 비율을 기반으로 부채꼴의 중심각이 정해짐.
df["Q1"].plot(kind = "pie")
```

```python
df["Q1"].plot(kind = "pie", autopct = "%1.1f%%")
```

```python
df["Q1"].plot(kind = "pie", autopct = "%1.1f%%", startangle = 90)
```

.plot 접근자를 사용하는 경우 "kind" 인자를 사용할 필요가 없다.

```python
df["Q1"].plot.bar()
```

```python
df["Q1"].plot.bar(color = "red")
```

```python
df["Q1"].plot.bar(color = "#FF0000")
```

```python
df["Q1"].plot.bar(color = "#FF0000", # 막대 면적 색상
                  edgecolor = "#000000", # 막대 테두리 색상
                  linewidth = 3) # 테두리 두께
```

DataFrame의 두 개 이상의 변수를 기반으로 선 그래프를 그리는 경우 각 변수의 값은 별도로 묶어서 그려주며 각 선의 정보는 범례에 표기된다.

```python
df[["Q1", "Q2"]].plot()
```

```python
df.plot.scatter(x = "Q1", y = "Q2", c = "red")
```

```python
df.plot.scatter(x = "Q1", y = "Q2", 
                c = ["#FF0000", "#00FF00", "#0000FF", "#000000"])
```

```python
df["color_num"] = [1, 1, 2, 2]
df.plot.scatter(x = "Q1", y = "Q2", 
                c = "color_num",
                cmap = "viridis")
```

```python
df.plot.scatter(x = "Q1", y = "Q2", 
                c = ["#FF0000", "#00FF00", "#0000FF", "#000000"],
                title = "title",
                xlabel = "xx",
                ylabel = "yy")
```
