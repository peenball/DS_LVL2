```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("mobiles.csv")
df.head(3)
```

```python
df.columns.unique()
```

```python
df["performance"] = df["ROM"]/32 + df["RAM"]/2 + (df["num_rear_camera"] + df["num_front_camera"]) + df["battery_capacity"]/1000
df.head(3)
```

```python
print(len(df))
print(np.mean(df["sales"]))
print(np.std(df["sales"]))
print(np.mean(df["sales"]) + 2 * np.std(df["sales"]))
```

```python
df["special"] = np.where(df["sales"] > np.mean(df["sales"]) + 2 * np.std(df["sales"]), 1, 0)
df.head(1)
```

```python
df_sub = df[df["special"]==1]
df_sub
```

```python
round(df_sub["performance"].mean(),2)
```
